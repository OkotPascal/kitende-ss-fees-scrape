#### Adding A Progress Bar

```python
scrapper_1 = sntwitter.TwitterSearchScraper('#Kitende')

tweets = [] 
n_tweets=500

for _, tweet in tqdm(enumerate(scrapper_1.get_items()), total=n_tweets):
    data = [tweet.date, 
        tweet.id, 
        tweet.rawContent, 
        tweet.user.username, 
        tweet.likeCount, 
        tweet.retweetCount]

    tweets.append(data)
    
    if _ > n_tweets:
        break

tweet_df = pd.DataFrame(tweets, columns=['tweet.date', 
                                          'tweet.id', 
                                          'tweet.rawContent', 
                                          'tweet.user.username', 
                                          'tweet.likeCount', 
                                          'tweet.retweetCount'])

tweet_df.to_csv('Kitende-tweets.csv', index=False)
```
