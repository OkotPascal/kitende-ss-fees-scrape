```python
scrapper = sntwitter.TwitterSearchScraper('#python')

tweets = []

for _, tweet in enumerate(scrapper.get_items()): # enumerate so we don't loose the data after running once
    data = [tweet.date, 
        tweet.id, 
        tweet.rawContent, 
        tweet.user.username, 
        tweet.likeCount, 
        tweet.retweetCount]

    tweets.append(data)
    
    if _ > 50:
        break
```

```python
len(tweets)
```

```python
tweets
```

```python
tweet_df = pd.DataFrame(tweets, columns=['tweet.date', 
                                          'tweet.id', 
                                          'tweet.rawContent', 
                                          'tweet.user.username', 
                                          'tweet.likeCount', 
                                          'tweet.retweetCount'])
tweet_df
```

```python
tweet_df.to_csv('python-tweets.csv', index=False)
```
